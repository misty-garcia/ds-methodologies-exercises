```python
feb = df["2019-02"].copy()
feb = feb.asfreq("D").fillna(0)
feb["shift1"] = feb.shift(1)
feb["shift3"] = feb.coffee_consumption.shift(3)
feb["shiftneg1"] = feb.coffee_consumption.shift(-1)
feb["deltacc"] = feb.coffee_consumption - feb.shift1
feb
```

```python
mar = df["2019-03"].copy().asfreq("D").fillna(0)
mar["shift1"] = mar.diff(1)
mar
```
